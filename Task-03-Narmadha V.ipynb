## **Project 3 : SQL Data Analysis** 

### Step 1 — Import Libraries

```python
import pandas as pd
import sqlite3
```

### Step 2 — Preview Dataset

```python
df = pd.read_excel("Cleaned_Dataset.xlsx")
df.head()
```

### Step 3 — Create SQL Database

```python
conn = sqlite3.connect("project3.db")
```

### Step 4 — Convert DataFrame to SQL Table

```python
df.to_sql("orders",conn,if_exists="replace",index=False)
```

### Step 5 — Verify Table

```python
query = """

SELECT * FROM orders
LIMIT 5

"""

pd.read_sql_query(query,conn)
```

- Displayed sample records to verify table creation.

### Step 6 — Count Total Orders

```python
query = """

SELECT COUNT(*) AS Total_Orders
FROM orders

"""

pd.read_sql_query(query,conn)
```

### Step 7 — Count Unique Customers

```python
query = """

SELECT COUNT(DISTINCT CustomerID) AS Customers
FROM orders

"""

pd.read_sql_query(query,conn)
```

### Step 8 — Count Products

```python
query = """

SELECT COUNT(DISTINCT Product) AS Products
FROM orders

"""

pd.read_sql_query(query,conn)
```

### Step 9 — Total Revenue

```python
query = """

SELECT SUM(TotalPrice) AS Revenue
FROM orders

"""

pd.read_sql_query(query,conn)
```

### Step 10 — Average Order Value

```python
query = """

SELECT AVG(TotalPrice) AS Average_Customer_Spending
FROM orders

"""

pd.read_sql_query(query,conn)
```

- Measured average customer spending per order.

### Step 11 — Average Quantity Purchased

```python
query = """

SELECT AVG(Quantity) AS Avg_Quantity
FROM orders

"""

pd.read_sql_query(query,conn)
```

- Calculated average units purchased per transaction i.e., ~ 3

### Step 12 — Revenue by Payment Method

```python
query = """

SELECT PaymentMethod, SUM(TotalPrice) AS Revenue
FROM orders

GROUP BY PaymentMethod
ORDER BY Revenue DESC

"""

pd.read_sql_query(query,conn)
```

- Compared revenue across payment methods.

### Step 13 — Top 10 Revenue Products

```python
query = """

SELECT Product, SUM(TotalPrice) AS Revenue
FROM orders

GROUP BY Product
ORDER BY Revenue DESC
LIMIT 10

"""

pd.read_sql_query(query,conn)
```

### Step 14 — Order Status Analysis

```python
query = """

SELECT OrderStatus, COUNT(*) AS Orders
FROM orders

GROUP BY OrderStatus
ORDER BY Orders DESC

"""

pd.read_sql_query(query,conn)
```

- Analyzed distribution of order completion status.

### Step 15 — Revenue by Referral Source

```python
query = """

SELECT ReferralSource, SUM(TotalPrice) AS Revenue
FROM orders

GROUP BY ReferralSource
ORDER BY Revenue DESC

"""

pd.read_sql_query(query,conn)
```

 - Measured revenue contribution from referral channels.

### Step 16 — Highest Value Orders

```python
query = """

SELECT OrderID, CustomerID, Product, Quantity, UnitPrice, TotalPrice, PaymentMethod
FROM orders

ORDER BY TotalPrice DESC
LIMIT 10

"""

pd.read_sql_query(query,conn)
```

### Step 17 — Products with Revenue Above 10000

```python
query = """

SELECT Product, SUM(TotalPrice) AS Revenue
FROM orders

GROUP BY Product
HAVING Revenue > 10000

"""

pd.read_sql_query(query,conn)
```

- Filtered products exceeding the revenue threshold.

### Step 18 — Customers with Highest Spending

```python
query = """

SELECT CustomerID, SUM(TotalPrice) AS Spending
FROM orders

GROUP BY CustomerID
ORDER BY Spending DESC
LIMIT 10

"""

pd.read_sql_query(query,conn)
```

### Step 19 — Close Connection

```python
# Closed SQL connection after analysis.
conn.close()
```
