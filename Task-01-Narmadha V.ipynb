## **Project 1 : Data Cleaning and Preparation**

### Step 1: Loading & Understanding the Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_excel("Dataset for Data Analytics.xlsx")
print(df.head())
print()
df.info()
print()
print(df.columns)
```

### Step 2: Handling missing values

```python
# Missing values
#print(df.isnull().sum())

# Fill missing coupon codes
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')
print(df['CouponCode'].head(10))

print(df.isnull().sum())
```

- Missing coupon values indicate no coupon applied, so replaced with 'No Coupon'.

### Step 3: Handling duplicates

```python
#check for duplicates
duplicate_rows = df[df.duplicated()]
print("Number of duplicate rows:", duplicate_rows.shape)

# View duplicate rows
print(duplicate_rows)

# Remove duplicates
df = df.drop_duplicates()
```

### Step 4: Validate Data Formats

```python
print(df.dtypes)

df['Date'] = pd.to_datetime(df['Date'])

print(df.dtypes)
```

```python
# Verify
print(df.isnull().sum())
print()

# Check numerical columns
print(df[['Quantity','UnitPrice','ItemsInCart','TotalPrice']].describe())
```

### Step 5: Verify Total Price Calculation

```python
# Create CalculatedPrice column
df['CalculatedPrice'] = df['Quantity'] * df['UnitPrice']

# Check mismatches
price_mismatch = (df['CalculatedPrice'] != df['TotalPrice']).sum()

print("Price Mismatches:", price_mismatch)

# Display sample rows
df[['Quantity', 'UnitPrice', 'TotalPrice', 'CalculatedPrice']].head()
```

### Step 6: Detect Outliers

```python
# Detect outliers in Quantity
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Quantity'] < lower) | (df['Quantity'] > upper)]

print("Number of Outliers:", len(outliers))
```

- Outliers were retained because high-value orders are valid business transactions.

### Step 7: Final Data Quality Validation

```python
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nDuplicate Order IDs:")
print(df['OrderID'].duplicated().sum())
```

### Step 8: Dataset Summary

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

### Step 9: Save the Cleaned Dataset

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)

print("Cleaned Dataset Saved Successfully")
```
